```python
datafile=pd.read_csv(r'stunted.csv')
table4= datafile.pivot_table(
    index= 'State/UT',
    values= ['stunted', 
             'wasted',
             'underweight'],

    aggfunc= 'mean'
)
    
table4.tail()

```

```python
merged_df=pd.merge(table4, df,on='State/UT',how='outer')
merged_df
mean_dropout_rate = merged_df['Dropout Rate (%)'].mean()  # Calculate the mean
merged_df['Dropout Rate (%)'] = merged_df['Dropout Rate (%)'].fillna(mean_dropout_rate)
merged_df

mean_wasting_rate = merged_df['wasted'].mean()  # Calculate the mean
merged_df['wasted'] = merged_df['wasted'].fillna(mean_wasting_rate)
merged_df

mean_underweight_rate = merged_df['underweight'].mean()  # Calculate the mean
merged_df['underweight'] = merged_df['underweight'].fillna(mean_underweight_rate)
merged_df

mean_stunting_rate = merged_df['stunted'].mean()  # Calculate the mean
merged_df['stunted'] = merged_df['stunted'].fillna(mean_stunting_rate)
merged_df.sample(5)
```
